# Small Things

Okay, so there are a few small things I noticed you could do differently, though your ways work!

The first is to do with the lower() and upper() methods which can be called on string objects. Have a look at the example code below - I've put your original code and then how I would do it (I've added notes, too). You should be able to see what I've done...

```python
answer1 = input()
# I love how you are confidently using the or operator...
if answer1 == ("Yes") or answer1 == ("yes"):
    print ("Ok, who? ")

# however, you could use methods to help you...
# ...the lower() method forces every character in the string stored in answer1 to be
# lower case!
if answer1.lower() == 'yes':
    print('Ok, who? ')
    
# you could of course force the characters to become upper case using the method upper()
if answer1.upper() == 'YES':
    print ("Ok, who?")
    
# but what if the player puts something like yep or yeah or yay or just y?
# you could use the startswith() method!
# you can add an argument to this method...
if answer1.startswith('y'):
    print ("Ok, who?")

# this would be a problem if the player puts something like Yes
# not really a problem!
# we can chain string methods on strings like so...
if answer1.lower().startswith('y'):
    print ("Ok, who?")
    
# I hope you can see what happens there...
# ...the lower() method forces the characters in the string stored in answer1 to become
# lower case...
# ...then the startswith('y') method checks if the string now starts with y
# you could of course do it this way...
if answer1.upper().startswith('Y'):
    print ("Ok, who?")
    
# problem solved!
# let's just hope that the player doesn't answer si, oui, sim or some other such word!
```

## Procedures and Functions

Okay, hopefully that made sense! The idea is that we break a larger program down into smaller programs (procedures and / or functions). This way of programming is called 'modular programming'. It makes our code easier to read and write!

Now you understand the idea, you should be able to see how it works in Python. Have a look at my version of your program below...

```python
def one():
    name = input("Ok, who? (Enter 1 if you don't know who!) ")
    if name == ("1"):
        doomed()
    else:
        two(name)

# so here I have used the keyword def which is used to define our own procedures
# and functions
# put simply, a function returns a value whereas a procedure does not
# however, from my experience, nobody really cares and there is mutual understanding anyway
# I'll use the term 'funtion' for the rest of this notebook
# procedures and functions are defined in Python using the keyword def

# notice that we have () after def and the name of the function
# we can pass values into our functions
# if we want to have a value passed into the function, this is where we name
# the value for the function - the name in () creates a variable which can be
# used inside the function

# I call the function called 'doomed' using doomed()
# this makes the program call 'doomed' so it will jump to the code in the earlier
# defined function 'doomed' - I defined it before I defined 'one'
# notice that I have not put any arguments (values I want to pass to 'doomed') inside ()

def doomed():
    print ("\nWelp, they're doomed")
    print ("\n*Bananaman flies away, deaf to your anguished yells...*")
    print ("\nYOU LOSE")
    quit()

# when I call the function called 'two', I do place a value into the () as an argument
# this is because I need the computer to use the value stored in the variable called 'name'
# in the function called 'two'
# I pass it kind of like a relay baton in a relay race!
# when I define 'two', I will create a variable which can store the value I am passing to
# it from 'one'
# this is the parameter I wrote about earlier

def two(name):
    print()
    print(name, " did you say? Where are they? ")
    location = input("\nEnter 1 if you haven't the foggiest where they are! ")
    if location == '1':
        doomed()
    elif location.lower() == 'here':
        three()
    else:
        four(name, location)
        
# so, I defined the function called 'two' and created a variable called 'name' in the ()
# this type of variable created in this way is called a parameter
# it will receive the value passed to it (like the relay baton!) from the argument
# given when it was called...one(name)
# this means the string stored in the variable 'name' in the function 'one' is now stored
# in the new variable called 'name' which has been created in the function 'two'

# notice that when I call the function called 'four', I pass two arguments to it...
# ...name and location
# 'name' is therefore just being passed along the functions as it is needed in them
# and this will happen with the string stored in 'location' as well

# how many parameters (variables) must I create when I define the function called 'four'

def four(name, location):
    print("\nSo, ", name, " is trapped in ", location,", huh? Come with me!")
    print ('''\n*You follow Bananaman into a secret lair full of strange gadgets.
You long to touch one. What do you do?*''')
    answer = input("\nA: you touch them, or B: you resist... ")
    if answer.upper() == 'A':
        five()
    elif answer.upper() == 'B':
        six(name, location)
    else:
        print("\nI'm sorry, Bananaman didn't quite get that!")
        
# yes, I needed to create two new variables (parameters) when I defined the function
# called 'four' because I am passing two values to it as arguments when I call it
# the number of parameters in the defined function must match the number of arguments
# passed to it when it is called

# so, how many parameters will there be when I define the function called 'six'

def six(name, location):
    print ("\n*You resist the temptation to touch them*")
    print ("\n*Soon Bananaman grabs you by the arm and flies off to ", location,
           ", where you meet ", name, "'s captor.*")
    meanie = input("\nWhat is the captor called? ")
    print ("\nWhat does", meanie, " look like? ")
    appearance = input('''\nA: they have sallow skin and an evil face...
B: they are short with a toothbrush moustache on their face...
C: they have green eyes and they are, in fact, an apple! ''')
    if appearance.upper() == 'A':
        seven(name, location, meanie)
    elif appearance.upper() == 'B':
        eight(name, location, meanie)
    elif appearance.upper() == 'C':
        nine(name, location, meanie)
    else:
        print("\nI'm sorry, Bananaman didn't quite get that...")
        
# yes, two, because I pass two values as arguments to it when I call it
# notice that I create a new variable inside this function and it is called 'meanie'
# this value (the bad guy's name) will be needed later in the story, so I need to
# pass it along (like another baton!) to new functions
# I pass it as an argument when I call the functions...
# ...seven(name, location, meanie)

# so, how many parameters will I create when I define the function called 'seven'

def seven(name, location, meanie):
    print ("\n*The sallow-skinned, evil faced ", meanie,
           "pulls a green ruler from their pocket and flings it at you. What do you do?*")
    choice = input("\nA: you try to catch the ruler, or B: you do nothing")
    if choice.upper() == ("A"):
        print ("\n*You raise your arm and catch the ruler, throwing it back at ",
               meanie, "'s face. They stumble to the ground, not yet defeated.*")
        print ("\n*Just as it seems all hope is lost, ", name,
               "comes to, and with a dazed grin, pummels ",meanie ,"into the ground...*")
        print ("\nCONGRATULATIONS! YOU WIN!")
        quit()
    elif choice.upper() == ("B"):
        print ("\n*The green ruler hits you square in the face, and you black out...*")
        print ("\nYOU LOSE")
        quit()
        
# yep - threedef seven(name, location, meanie):
    print ("\n*The sallow-skinned, evil faced ", meanie,
           "pulls a green ruler from their pocket and flings it at you. What do you do?*")
    choice = input("\nA: you try to catch the ruler, or B: you do nothing")
    if choice.upper() == ("A"):
        print ("\n*You raise your arm and catch the ruler, throwing it back at ",
               meanie, "'s face. They stumble to the ground, not yet defeated.*")
        print ("\n*Just as it seems all hope is lost, ", name,
               "comes to, and with a dazed grin, pummels ",meanie ,"into the ground...*")
        print ("\nCONGRATULATIONS! YOU WIN!")
        quit()
    elif choice.upper() == ("B"):
        print ("\n*The green ruler hits you square in the face, and you black out...*")
        print ("\nYOU LOSE")
        quit()
```
